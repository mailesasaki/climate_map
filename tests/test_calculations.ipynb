```python
def calc_wind_chill(tasmin, sfcWind):

    tasmin_F = tasmin.tasmin * 9/5 - 459.67 # Convert to Fahrenheit
    sfcWind_mph = 0.44704 * sfcWind.sfcWind # Convert to mph
    # From NWS https://www.weather.gov/safety/cold-wind-chill-chart
    
   # Note: The Wind Chill Temperature is defined only for temperatures at or below 50°F and wind speeds above 3 mph.
    tasmin_F = tasmin_F.where(tasmin_F<50)

    sfcWind_mph = sfcWind_mph.interp(lat=tasmin_F.lat, lon=tasmin_F.lon) # Ask about best way to interpolate?
    sfcWind_mph = sfcWind_mph.where(sfcWind_mph>3)
    #tasmin_aligned, sfcWind_aligned = xr.align(tasmin_F, sfcWind_mph)
    
    wind_chill = 35.74 + 0.6215*tasmin_F - 35.75*(sfcWind_mph**0.16) + 0.4275*tasmin_F*(sfcWind_mph**0.16)
    
    return wind_chill, tasmin_F, sfcWind_mph
```

```python
wind_chill, tasmin_F, sfcWind_mph = calc_wind_chill(tasmin, sfcWind)
```
